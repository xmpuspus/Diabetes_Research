```python
pul_claims =  op_claims.merge(ip_claims, how='outer', on=['RandomMbrID', 'FULL_DT'])
pul_claims = pul_claims.merge(dr_claims, how='outer', on=['RandomMbrID', 'FULL_DT'], suffixes=('_x1', '_y1'))
pul_claims = pul_claims.merge(rx_claims, how='outer', on=['RandomMbrID', 'FULL_DT'], suffixes=('_x2', '_y2'))
pul_claims.head()
```

```python
pul_claims = pul_claims[['RandomMbrID', 'FULL_DT'] + [i for i in pul_claims.columns if 'Allw_Amt' in i]]
pul_claims.head()
```

```python
pul_claims['FULL_DT'] = pd.to_datetime(pul_claims['FULL_DT'])
pul_claims['Year'] = pul_claims.FULL_DT.dt.year
pul_claims['Month'] = pul_claims.FULL_DT.dt.month
member_claims = pul_claims.merge(members, on='RandomMbrID', how='inner')
member_claims['Allw_Amt'] = member_claims[[i for i in pul_claims.columns if 'Allw_Amt' in i]].sum(axis=1)
member_claims.drop([i for i in pul_claims.columns if 'Allw_Amt' in i], axis=1, inplace=True)
member_claims = member_claims.groupby(['RandomMbrID', 'Age', 'Gdr_Cd', 'Zip3', 'FULL_DT', 'Year', 'Month'], as_index=False).sum()
member_claims.head()
```

```python
op_ip_claims = pd.merge(op_claims, ip_claims, how='outer', left_on=['RandomMbrID', 'FULL_DT'], right_on=['RandomMbrID', 'FULL_DT'])
all_claims = pd.merge(op_ip_claims, dr_claims, how='outer', left_on=['RandomMbrID', 'FULL_DT'], right_on=['RandomMbrID', 'FULL_DT'])
all_claims[list(all_claims.loc[:, 'Allw_Amt_x':])] = all_claims[list(all_claims.loc[:, 'Allw_Amt_x':])].astype(float).fillna(0)
all_claims = all_claims[(all_claims['Allw_Amt_x'] >= 0) & (all_claims['Allw_Amt_y'] >= 0) & (all_claims['Allw_Amt'] >= 0)]
all_claims['Allw_Amt_Sum'] = all_claims[['Allw_Amt_x', 'Allw_Amt_y', 'Allw_Amt']].sum(axis=1)
all_claims.drop(['Allw_Amt_x', 'Allw_Amt_y', 'Allw_Amt'], axis=1, inplace=True)
all_claims['Month'] = all_claims['FULL_DT'].astype(str).str[5:7]

def quarter(row):
    if row['Month'] == '01' or row['Month'] =='02' or row['Month'] == '03':
        return 1
    elif row['Month'] == '04' or row['Month'] =='05' or row['Month'] == '06':
        return 2
    elif row['Month'] == '07' or row['Month'] =='08' or row['Month'] == '09':
        return 3
    elif row['Month'] == '10' or row['Month'] =='11' or row['Month'] == '12':
        return 4

all_claims['Quarter'] = all_claims.apply(lambda row : quarter(row), axis=1)
all_claims.head()
```

```python
members_claims = pd.merge(members, all_claims, how='inner', left_on='RandomMbrID', right_on='RandomMbrID')
members_claims.head()
```

```python
members_zip_month = members_claims.groupby(['Zip3', 'Month'], sort=True, as_index=False).mean()
members_zip_month
```

```python
members_zip_quarter = members_claims.groupby(['Zip3', 'Quarter'], sort=True, as_index=False).mean()
members_zip_quarter
```
